2. Cree el entorno CartPole

```python
import gym

env = gym.make('CartPole-v1')
```

3. Definición de las redes en línea y de destino


```python
import torch.nn as nn

#ff
class DQNNetwork(nn.Module):
    def __init__(self, input_dim, output_dim, hidden_dim=64):
        super(DQNNetwork, self).__init__()
        self.fc1 = nn.Linear(input_dim, hidden_dim)
        self.fc2 = nn.Linear(hidden_dim, output_dim)

    def forward(self, x):
        x = torch.relu(self.fc1(x))
        x = self.fc2(x)
        return x

#Creamos las dos redes. Tenemos la online y la de destino
online_network = DQNNetwork(env.observation_space.shape[0], env.action_space.n)
target_network = DQNNetwork(env.observation_space.shape[0], env.action_space.n)
target_network.load_state_dict(online_network.state_dict())
```

4. Establecimiento de hiperparámetros:


```python
# Hiperparámetros
num_episodes = 1000
batch_size = 64
gamma = 0.99
epsilon = 1.0
epsilon_min = 0.01
epsilon_decay = 0.995
target_update_frequency = 5
```

5. Selección de acciones épsilon-greedy:


```python
import numpy as np

def epsilon_greedy_action(network, state, epsilon):
    if np.random.rand() < epsilon:
        return env.action_space.sample()
    else:
        with torch.no_grad():
            return torch.argmax(network(torch.tensor(state).float())).item()
```

6. Definición de la reproducción de la experiencia (experience replay):


```python
from collections import deque
import random

class ReplayBuffer:
    def __init__(self, capacity):
        self.buffer = deque(maxlen=capacity)

    def push(self, state, action, reward, next_state, done):
        experience = (state, action, reward, next_state, done)
        self.buffer.append(experience)

    def sample(self, batch_size):
        return random.sample(self.buffer, batch_size)

    def __len__(self):
        return len(self.buffer)

replay_buffer = ReplayBuffer(10000)
```

7. Ciclo de entrenamiento:


```python
import torch.optim as optim
import numpy as np
import torch



optimizer = optim.Adam(online_network.parameters(), lr=0.001)

for episode in range(num_episodes):
    state = env.reset()
    episode_reward = 0
    done = False

    while not done:
        action = epsilon_greedy_action(online_network, state, epsilon)
        next_state, reward, done, _ = env.step(action)
        replay_buffer.push(state, action, reward, next_state, done)

        episode_reward += reward
        state = next_state

        if len(replay_buffer) > batch_size:
            experiences = replay_buffer.sample(batch_size)
            states, actions, rewards, next_states, dones = zip(*experiences)

            states = torch.tensor(states).float()
            actions = torch.tensor(actions).long().unsqueeze(-1)
            rewards = torch.tensor(rewards).float().unsqueeze(-1)
            next_states = torch.tensor(next_states).float()
            dones = torch.tensor(dones).float().unsqueeze(-1)

            current_q_values = online_network(states).gather(1, actions)
            next_q_values = target_network(next_states).max(dim=1)[0].unsqueeze(-1)
            target_q_values = rewards + gamma * next_q_values * (1 - dones)

            loss = nn.functional.mse_loss(current_q_values, target_q_values)

            optimizer.zero_grad()
            loss.backward()
            optimizer.step()

        if episode % target_update_frequency == 0:
            target_network.load_state_dict(online_network.state_dict())

    epsilon = max(epsilon_min, epsilon_decay * epsilon)
    print(f"Episode {episode}: Total Reward: {episode_reward}")
```

```python
import matplotlib.pyplot as plt

rewards = []

for episode in range(num_episodes):
    rewards.append(episode_reward)

plt.plot(rewards)
plt.ylabel('Reward acumulado')
plt.xlabel('Episodio')
plt.title('Progreso del entrenamiento del agente DQN')
plt.show()

test_episodes = 10
test_rewards = []

for episode in range(test_episodes):
    state = env.reset()
    episode_reward = 0
    done = False

    while not done:
        action = torch.argmax(online_network(torch.tensor(state).float())).item()
        next_state, reward, done, _ = env.step(action)
        episode_reward += reward
        state = next_state

    test_rewards.append(episode_reward)

average_test_reward = sum(test_rewards) / test_episodes
print(f"Recompensa promedio sobre {test_episodes} episodios de prueba: {average_test_reward}")

# Cerrando el entorno después de todo
env.close()
```

Defina en qué consiste y en qué clase de problemas se pueden usar cada uno de los siguientes acercamientos en
Deep Reinforcement Learning
1. Proximal Policy Optimization

Descripción: PPO es un método de optimización de políticas que modifica la función objetivo para entrenar una política actualizada, de manera que no se desvíe mucho de una política anterior. Lo hace mediante la introducción de un término de recorte en la función objetivo, lo que evita actualizaciones excesivamente grandes.
Aplicaciones: PPO es útil para una variedad de problemas de RL, incluidos entornos con espacio de acción continuo y discreto. Ha sido ampliamente adoptado en la comunidad RL debido a su robustez y facilidad de implementación, y se ha utilizado con éxito en entornos como los de OpenAI Gym y robótica.

2. Deep Deterministic Policy Gradients (DDPG)

Descripción: DDPG es un algoritmo que combina ideas de Deep Q-Learning y política determinista basada en gradientes. Es un método actor-crítico, donde el actor produce una acción determinista y el crítico evalúa esa acción. DDPG utiliza memoria de repetición y redes objetivo para estabilizar el entrenamiento, similar a DQN.

Aplicaciones: DDPG es particularmente adecuado para entornos con espacios de acción continua, lo que lo hace útil para aplicaciones como el control de robótica y tareas que requieren una gran precisión en las acciones.

3. Trust Region Policy Optimization (TRPO)

Descripción: TRPO es un algoritmo de optimización de políticas que garantiza que las actualizaciones a la política no cambien la política anterior más allá de un cierto límite definido. Lo hace optimizando una función objetivo sujeto a una restricción que limita la variación de la política.

Aplicaciones: Similar a PPO, TRPO es adecuado para una amplia variedad de problemas de RL. Sin embargo, TRPO puede ser más complicado de implementar y computacionalmente más intensivo que PPO debido a la necesidad de realizar optimización con restricciones.

4. Asynchronous Advantage Actor-Critic (A3C)

Descripción: A3C es un método de aprendizaje por refuerzo que utiliza múltiples agentes que se ejecutan en paralelo en múltiples instancias del entorno. Estos agentes actualizan de manera asincrónica una política global. El método combina ideas de "Advantage Actor-Critic" (una variante de métodos actor-crítico) y entrenamiento asíncrono.

Aplicaciones: A3C es adecuado para una amplia gama de problemas de RL y ha mostrado un rendimiento impresionante en tareas como el conjunto de juegos de Atari. Debido a su enfoque paralelo, puede ser más eficiente en términos de tiempo de entrenamiento que algunos otros métodos, dependiendo de la infraestructura disponible.
